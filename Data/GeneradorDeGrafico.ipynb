```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
all = pd.read_excel(r"Datos.xlsx")
all2 = pd.DataFrame(all)

fecha = all2["fecha"].tolist()
precioCasa = all2["preciocasa"].tolist()
precioApartamento = all2["precioapartamento"].tolist()
```

```python
fig, ax = plt.subplots()
ax.plot(fecha, precioCasa, label="Casas")
ax.plot(fecha, precioApartamento, label="Apartamentos")
ax.axis((0, len(fecha), 0, max(precioCasa)*1.3))
ax.xaxis.set_major_locator(plt.MaxNLocator(7))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: 'US$ ' + '{:,.0f}'.format(x)))
plt.legend()
plt.show()
```
